```python

import pandas as pd
datas=pd.read_csv("/workspaces/Proyecto_Final_Grupo_9/pandas/Quiz Cálculo 1- Gobierno 22 de Octubre(1-18).csv")
import matplotlib.pyplot as plt
valores=[datas["NF"].quantile(0.10)
,datas["NF"].quantile(0.20),datas["NF"].quantile(0.30),datas["NF"].quantile(0.40),datas["NF"].quantile(0.50),datas["NF"].quantile(0.60),datas["NF"].quantile(0.70),datas["NF"].quantile(0.80),datas["NF"].quantile(0.90)]
etiquetas=["P10","P20","P30","P40","P50","P60","P70","P80","P90"]
plt.bar(etiquetas,valores)
plt.xlabel("Percentiles")
plt.ylabel("Valores")
plt.title("Distribución de Percentiles de la Nota Final")
plt.show()   
```
