```python
import matplotlib.pyplot as plt
import pandas as pd
datas = pd.read_csv("/workspaces/Proyecto_Final_Grupo_9/pandas/Quiz CÃ¡lculo 1- Gobierno 22 de Octubre(1-18).csv")
```

```python
def promedios(nombre_archivo, nombre_columna):
    if nombre_columna not in nombre_archivo.columns:
        raise  ValueError('Error al procesar la columna')
    def prom(x):
        p = 0
        for i in x:
            p += i
        promedio = p/len(x)
        return promedio
    return prom(nombre_archivo[nombre_columna])
values=[promedios(datas,"CP1"),promedios(datas,"CP2"),promedios(datas,"CP3"),promedios(datas,"NF")]
label=["prom CP1","prom CP2","prom CP3","prom NF"]
plt.bar(label,values)
plt.xlabel("Promedios")
plt.ylabel("Valores")
plt.title("Promedios de CP1,CP2,CP3,NF")
plt.show()
```
