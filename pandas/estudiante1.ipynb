```python
import pandas as pd
```

```python
datas=pd.read_csv("/workspaces/Proyecto_Final_Grupo_9/pandas/Quiz Cálculo 1- Gobierno 22 de Octubre(1-18).csv")
```

```python
datas
```

```python
def polinomio1(x):
  o = x*(101)**2 + (2*101) + 23
  t = x*(100)**2 + (2*100) + 23
  e = (o-t)/(101-100)
  return e
datas["P2"] = datas["N"].apply(polinomio1)



def calificacion1(x):
      if x["P2"] == x["Sea la función de costo C(x) = n*x^2 + 2*x + 23.\xa0\nHallar el costo promedio de producir\xa0 de 100 a 101 unidades (Poner dos decimales)\n"]:
        return 5
      else:
        return 1
datas["Calificaciones P2"] = datas.apply(calificacion1, axis = 1)
datas
```

```python

datas=datas.drop(["Datos Reales 2","Calificacion 2","Calificaciones P3"],axis=1)
```

```python
def costo_marginal(x):
    c=(2*100*x) + 2
    return float(c)

datas["P3"]=datas["N"].apply(costo_marginal)
datas
```

```python
datas.iat[17,3]=float(datas.iat[17,3].replace(",","."))
```

```python

def califica3(x):
     if x["P3"]==float(x["Sea la función de costo C(x) = n*x^2 + 2*x + 23.\xa0\nUsar el costo marginal para aproximar el costo promedio entre x= 100 y x =101 (Poner dos decimales)"]):
        return 5
     else:
        return 1
datas["Calificaciones P3"]=datas.apply(califica3,axis=1)
datas
```

```python
datas=datas.rename(columns={"Calificaciones P2":"CP2","Calificaciones P3":"CP3")
datas
```

```python
import matplotlib.pyplot as plt
def sd(data, columna):
    try:
        return data[columna].std()
    except KeyError:
        return 'Error al procesar la columna'
    except Exception as e:
        return 'Error al procesar la columna'
values=[sd(datas,"CP1"),sd(datas,"CP2"),sd(datas,"CP3"),sd(datas,"NF")]
label=["desv CP1","desv CP2","desv CP3","desv NF"]
plt.bar(label,values)
plt.xlabel("Desviaciones estandar")
plt.ylabel("Valores")
plt.title("Distribucion de las desviaciones estándar de CP1,CP2,CP3,NF")
plt.show()
```
