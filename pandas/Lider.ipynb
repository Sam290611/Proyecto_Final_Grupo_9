```python
def suma(x):
  p = 0
  for i in str(x):
    p += int(i)
  return p
```

```python
datas["N"] = datas["Número de documento"].apply(suma)
datas
```

```python
def polinomio(x):
  m = x*(105)**2 + (2*105) + 23
  v = x*(100)**2 + (2*100) + 23
  h = (m-v)/(105-100)
  return h
datas["Datos Reales"] = datas["N"].apply(polinomio)
datas
```
